### Resultados



En este repositorio se adjunta un documento de Excel el cual contiene todos los datos de los experimentos a detalle, separados por distintas ventanas.

Los resultados obtenidos son los esperados. El SSD es el disco más rápido, seguido por el HDD y por último el HDD externo. 
A continuación dos tablas que contienen los datos más importantes, la Velocidad y el tiempo de ejecución para cada uno de los Block Size dados.

Se puede ver una tendencia clara respecto al tamaño de los Block Sizes. <br>
Mientras más grande es el tamaño del BS, máyor será el tiempo de ejecución del comando pero de la misma manera, mayor será la velocidad de la memoria del disco. Esto se debe a que cuando la muestra es muy pequeña, la ejecución del comando acaba tan rápido que el disco pueda usar todas sus capacidades.

Por otra parte, se puede ver que cuando no se usa el oflag, las velocidades tienden a ser menores, esto se debe a que el flag le da ciertos poderes al comando 'dd'. <br>
El primero de estos poderes es que permite que el kernel acceda directamente a los datos, sin necesidad de esperar a un buffer. Esto hace que la velocidad sea más rápida. <br>
No se le da el trabajo intermedio al kernel para que lo realice sino que solo ejecuta el código en ese momento y ya.
